## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


````python
T = identity_matrix[rows]
print("First row : ",T[0])
print("Second row : ",T[0])
print("Third row : ",T[2])
print("Based on the information provided in the previous sections, it can be observed that the T matrix is essentially an identity matrix that has been shuffled according to the rows array generated using the np.random.shuffle function.")
````

````python
print("Shape of T matrix : ", T.shape)
````

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see.
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

````python
transformed_image = T @ img_average_channel
print("Shape of the resultant matrix : ", transformed_image.shape)
````

````python
plt.imshow(transformed_image)
plt.show()
````

````python
print("Based on the image displayed above, it is evident that the image has undergone a scrambling process. ")
print("Specifically, the rows of the original image have been rearranged according to the predefined array 'rows' that we generated.")
````

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

````python
print("To determine the invertibility of matrix T, we will assess its determinant:")

print("1)If the determinant equals 0, then T is non-invertible.")
print("2)If the determinant is not equal to 0, then T is invertible.")
````

````python
invertible = np.linalg.det(T)
if invertible == 0:
    print("T is non-invertible")
else:
    print("T is invertible")
````

````python
T_INV = np.linalg.inv(T)
old = T_INV @ transformed_image
plt.imshow(old)
plt.show()
````

## Part 8: Extra Points

````python
print("If we wish to apply the same operations to the columns of the original image, we can achieve this by simply transposing the T matrix. ")
print("This transposed matrix will then represent the reordering of the columns of the original image matrix.")
````

````python
column_t = img_average_channel @ T.T
plt.imshow(column_t)
plt.show()
````

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

````python
print("If we intend to perform both column and row scrambling simultaneously, we can achieve this by multiplying both the T and its transpose (T transposed) matrices with the img_average_channel. ")
print("This combined operation will yield the desired result.")
````

````python
combine_t = T @ img_average_channel @ T.T
plt.imshow(combine_t)
plt.show()
````

````python
print("To restore the original image, we will perform the inverse of the matrix multiplication operation. ")
print("Specifically, we will multiply the Tinv matrix by its transpose (Tinv.transpose) with the combine_transformed image.")
````

````python
original_image = T_INV @ combine_t @ T_INV.T
plt.imshow(original_image)
plt.show()
````
